# Data Sampling

```python
y = wine_data.pop('Country_Code')
y = pd.DataFrame(y)
y.head()
```

```python
x = wine_data
x
```

```python
df1 = x.drop(['Winery', 'Wine', 'Primary_Grape', 'Natural', 'Style'], axis = 1)
```

```python
df1
```

```python
y['Country_Code'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(random_state=312)
X_ad, Y_ad = oversample.fit_resample(df1, y)
print(Y_ad.value_counts())
```

```python
X_ad
```

```python
len(Y_ad)
```

```python
result = pd.merge(X_ad, x, how='left')
```

```python
X_ad = result
```

```python
# train:val:test = 6:2:2
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_ad, Y_ad, test_size=0.2, random_state=1) 
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```
